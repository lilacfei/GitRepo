```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.multiclass import OneVsRestClassifier
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
train_data = pd.read_csv('test.csv')
x = train_data.iloc[:, 2:] #x是给定数据
```

```python
from sklearn.manifold import TSNE
```

```python
tsne = TSNE(n_components=2, learning_rate=100).fit_transform(x)
```

```python
pd.DataFrame(tsne).describe()
```

```python
tsne = np.array(tsne)
```

```python
plt.figure(figsize=(8, 6))
plt.subplot(111)
plt.scatter(tsne[:, 0], tsne[:, 1])
plt.show()
```

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN()
db_x=dbscan.fit_predict(tsne)
pd.DataFrame(db_x).describe()
```

```python
plt.scatter(tsne[:,0], tsne[:,1], c=db_x)
plt.show()
plt.savefig('dbscan.png')
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans()
km_x=kmeans.fit_predict(tsne)
pd.DataFrame(km_x).describe()
```

```python
plt.scatter(tsne[:,0], tsne[:,1], c=km_x)
plt.show()
plt.savefig('kmeans.png')
```

```python
from sklearn import mixture
gmm = mixture.GaussianMixture()
gm_x=gmm.fit_predict(tsne)
pd.DataFrame(gm_x).describe()
```

```python
plt.scatter(tsne[:,0], tsne[:,1], c=gm_x)
plt.show()
plt.savefig('GMM.png')
```

```python
from sklearn.cluster import AgglomerativeClustering
hirc = AgglomerativeClustering()
hir_x=hirc.fit_predict(tsne)
pd.DataFrame(hir_x).describe()
```

```python
plt.scatter(tsne[:,0], tsne[:,1], c=hir_x)
plt.show()
plt.savefig('层次聚类.png')
```
